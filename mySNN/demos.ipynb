```python
T1, T2, h = 0, 100, 0.1
clk = np.arange(T1, T2, h)  
in1 = dummyNode()
out1 = regularSpiking()
s = Synapse(in1, out1)

for t in clk:
    in1.step(t)
    out1.step(t)
    s.step(t)

plt.figure()
plt.plot(clk,in1.iMemory,label = 'offset=6')
plt.plot(clk,in1.vMemory,label='neuron1')
plt.plot(clk,s.net_memory,label='postsynaptic potential')
plt.plot(clk,out1.vMemory,label='neuron2')
plt.legend()
plt.show()
```

```python
import SNNconnectivity as c
from SNNneuron import Neuron, regularSpiking

import random
import matplotlib.pyplot as plt
import numpy as npim
```

Tuning curves randomly assigned to neurons based on f( +/- x) = sqrt( x - c) c is a constant inthe range of input amplitude

```python
r = regularSpiking()
T1, T2, h = 0, 1000, 0.1
clk = np.arange(T1, T2, h)  

scale = 100
I1, I2 , Ih = (-1 *scale, 1 *scale, 0.01*scale)
offset_list = np.arange(I1, I2, Ih)
neuron_list = [regularSpiking() for k in range(10)]
tfs = np.arange(I1, I2, (I2 -I1)/len(neuron_list))
inputmem = []
fr = []
im = []
temp = []
temp2 = []

for j,r in enumerate(neuron_list):
    r.tuning_factor = tfs[j]
    for i in offset_list: # runs through each input level
        r.clear_mem()
        r.rest_I = [i]
        for t in clk: # runs each neuron
            r.step(t)
        temp.append(r.get_firing_rate(T2/1000))
        temp2.append(r.iMemory[-1])
    fr.append(temp)
    im.append(temp2)
    temp = []
    temp2 = []
    
```

```python
for i,r in enumerate(neuron_list):
    plt.plot(offset_list, fr[i],label = i)
plt.legend()
plt.title('tuning curves of neurons')
plt.xlabel('input range')
plt.ylabel('frequency response Hz')
plt.show()


plt.show()
```
